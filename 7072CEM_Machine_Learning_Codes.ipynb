```python
import pandas as pd
import matplotlib.pyplot as plt
from imblearn.over_sampling import RandomOverSampler

# Load the dataset
data = pd.read_csv("online_shoppers_intention.csv")

# Separate the features (X) and the target variable (y)
X = data.drop('Revenue', axis=1)
y = data['Revenue']

# Count the occurrences of each class before oversampling
class_counts_before = y.value_counts()

# Apply Random Oversampling
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Create a DataFrame with the resampled data
resampled_data = pd.concat([X_resampled, y_resampled], axis=1)

# Count the occurrences of each class after oversampling
class_counts_after = resampled_data['Revenue'].value_counts()

# Plot the class distribution before and after oversampling
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot before oversampling
axs[0].bar(class_counts_before.index, class_counts_before.values)
axs[0].set_xlabel('Class')
axs[0].set_ylabel('Count')
axs[0].set_title('Class Distribution Before Oversampling')

# Plot after oversampling
axs[1].bar(class_counts_after.index, class_counts_after.values)
axs[1].set_xlabel('Class')
axs[1].set_ylabel('Count')
axs[1].set_title('Class Distribution After Oversampling')

plt.tight_layout()
plt.show()
```
